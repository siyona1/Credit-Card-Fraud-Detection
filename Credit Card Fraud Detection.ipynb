```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
data = pd.read_csv("creditcard.csv")
```

```python
data.head(10)
```

```python
data['Class'].value_counts()
```

```python
normal=data[data.Class==0]
```

```python
fraud=data[data.Class==1]
```

```python
normal.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
data.groupby('Class').mean()
```

```python
normal_sample=normal.sample(n=492)
```

```python
new_file=pd.concat([normal_sample,fraud],axis=0)
```

```python
new_file.head(10)
```

```python
new_file['Class'].value_counts()
```

```python
new_file.groupby('Class').mean()
```

```python
X=new_file.drop(columns='Class',axis=1)
```

```python
Y=new_file['Class']
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,stratify=Y,random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import make_pipeline

model = LogisticRegression(max_iter=1000)

pipeline = make_pipeline(StandardScaler(), model)

pipeline.fit(X_train, Y_train)
```

```python
X_train_prediction=pipeline.predict(X_train)
```

```python
training_data_acuracy=accuracy_score(X_train_prediction,Y_train)*100
```

```python
print(f"Training Accuracy: {training_data_acuracy}%")
```

```python
X_test_prediction=pipeline.predict(X_test)
```

```python
test_data_accuracy=accuracy_score(X_test_prediction,Y_test)*100
```

```python
print(f"Testing Accuracy: {test_data_accuracy}%")
```
